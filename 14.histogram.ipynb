```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=(80,100,125,150),rwidth=0.9,color="red",histtype='step')
```

```python
plt.xlabel('Sugar level',size=18)
plt.ylabel('No of patient',size=18)
plt.title('Blood Sugar Chart',size=21)
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],color=['black','pink'],label=['MEN','GIRL'])

plt.legend()
```
