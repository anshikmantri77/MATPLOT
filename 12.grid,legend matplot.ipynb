```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=2,3,4,5,6
y=23,24,25,34,35
```

```python
plt.plot(x,y,'rD--',linewidth=4)
```

```python
plt.plot(x,y,color='green',marker='+',linestyle='dotted',linewidth=4)
```

```python
plt.plot(x,y,color='green',alpha=0.1)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[40,51,52,48,47,49,46]
min_t=[40,42,40,44,33,35,37]
avg_t=[40,48,48,46,40,42,41]
```

```python
plt.xlabel('DAY',size=15)
plt.ylabel('TEMP',size=15)
plt.title('WEATHER',size=18)
plt.plot(days,max_t,color='red',linestyle='dotted',label="Max")
plt.plot(days,min_t,color='green',linestyle='dashdot',label="Min")
plt.plot(days,avg_t,color='blue',linestyle='dotted',label="Avg")

plt.legend(loc="best",shadow='true',fontsize='small')

plt.grid()
```
