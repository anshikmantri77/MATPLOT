```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,10,15,8]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company",size=15)
plt.ylabel("Revenue",size=15)
plt.title("US stocks",size=18)
plt.bar(xpos,revenue,label='Revenue',color='orange')
plt.bar(xpos,profit,label='Profit',color='green')

plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company",size=15)
plt.ylabel("Revenue",size=15)
plt.title("US stocks",size=18)
plt.bar(xpos-0.1,revenue,width=0.4,label='Revenue',color='orange')
plt.bar(xpos+0.1,profit,width=0.4,label='Profit',color='green')

plt.legend()
```

```python
plt.xticks(xpos,company)

plt.title("US stocks",size=18)
plt.barh(xpos+0.1,revenue,label='Revenue',color='orange')
plt.barh(xpos-0.1,profit,label='Profit',color='green')

plt.legend()
```
